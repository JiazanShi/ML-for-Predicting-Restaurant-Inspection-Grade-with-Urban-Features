## DSNY data

Food scrap drop-off locations are for NYC resident to drop their food scrap, we use this data to reflect the distribution of urban waste management resources.

```python
Dropoff = pd.read_csv('Food_Scrap_Drop-Off_Locations_in_NYC.csv')
Dropoff.head()
```

```python
Dropoff = Dropoff.loc[~Dropoff['Location Point'].isnull()]
Dropoff = Dropoff.loc[Dropoff['Open_Month'] == 'Year Round']
```

```python
import geopandas as gpd
from shapely.geometry import Point

geometry = [Point(xy) for xy in zip(Dropoff.Longitude, Dropoff.Latitude)]
geodata = gpd.GeoDataFrame(Dropoff[:],geometry=geometry,crs={'init': 'epsg:4326'})
geodata.head()
```

```python
zips = gpd.read_file('ZIP_CODE/ZIP_CODE_040114.shp')
zips.head()
```

```python
zips.to_crs(4326,inplace=True)
```

```python
DOLocation = gpd.sjoin(geodata,zips,how='left',op='within')
DOLocation.head()
```

```python
DOLocation = DOLocation.loc[~DOLocation['ZIPCODE'].isnull()]
```

```python
DOagg = DOLocation.groupby(['ZIPCODE']).agg({'Object ID':'count'}).reset_index().rename(columns = {'Object ID':'num_foodDO'})
DOagg.head()
```

```python
DOagg.to_csv('dropoff.csv',index=False)
```

## Median Income

```python
income = pd.read_csv('Median income/ACSST5Y2021.S1903-Data.csv')
#S1903_C03_001E': Estimate!!Median income (dollars)!!HOUSEHOLD INCOME BY RACE AND HISPANIC OR LATINO ORIGIN OF HOUSEHOLDER!!Households
income = income[['NAME','S1903_C03_001E']]
income = income.rename(columns = {'S1903_C03_001E':'median income'})
income.head()
```

```python
income['NAME'] = income['NAME'].apply(lambda x: x.split()[1])
income = income.rename(columns = {'NAME':'ZIP'})
income.head()
```

```python
income['ZIP'] = income['ZIP'].astype(int)
```

```python
def remove_punc(docs):
    punctuations = '''!()[]{};:"\,<>/?@#$%^&*_~-+'''
    docs2=[];
    i=0;
    for sent in docs:
        s=""
        for x in sent:
            if(x not in punctuations):
                s=s+x;
        docs2.append(s);


    docs=docs2;
    docs2=[]
    return docs
```

```python
#manhattan median
Man = income.loc[(income['ZIP']>10000) & (income['ZIP']<10300)]
Man = Man.loc[Man['median income']!='-']
Man['median income'] = remove_punc(Man['median income'])
Man_median = Man['median income'].median()
```

```python
#staten island median
SI = income.loc[(income['ZIP']>10300) & (income['ZIP']<10400)]
SI = SI.loc[SI['median income']!='-']
SI['median income'] = remove_punc(SI['median income'])
SI_median = SI['median income'].median()
```

```python
#Bronx median
Br = income.loc[(income['ZIP']>10400) & (income['ZIP']<10500)]
Br = Br.loc[Br['median income']!='-']
Br['median income'] = remove_punc(Br['median income'])
Br_median = Br['median income'].median()
```

```python
#BK
BK = income.loc[(income['ZIP']>=11200) & (income['ZIP']<11300)]
BK = BK.loc[BK['median income']!='-']
BK['median income'] = remove_punc(BK['median income'])
BK_median = BK['median income'].median()
```

```python
#queens
Q = income.loc[((income['ZIP']>=11000) & (income['ZIP']<11200)) | ((income['ZIP']>=11300) & (income['ZIP']<11700))]
Q = Q.loc[Q['median income']!='-']
Q['median income'] = remove_punc(Q['median income'])
Q_median = Q['median income'].median()
```

```python
print(Man_median)
print(SI_median)
print(Br_median)
print(BK_median)
print(Q_median)
```

```python
for i in income.loc[income['median income']=='-'].index:
    if income['ZIP'].iloc[i] > 10000 and income['ZIP'].iloc[i] < 10300:
        income['median income'].iloc[i] = Man_median
    elif income['ZIP'].iloc[i] > 10300 and income['ZIP'].iloc[i] < 10400:
        income['median income'].iloc[i] = SI_median
    elif income['ZIP'].iloc[i] > 10400 and income['ZIP'].iloc[i] < 10500:
        income['median income'].iloc[i] = Br_median
    elif income['ZIP'].iloc[i] > 11200 and income['ZIP'].iloc[i] < 11300:
        income['median income'].iloc[i] = BK_median
    else:
        income['median income'].iloc[i] = Q_median
```

```python
income.to_csv('medianincome.csv',index=False)
```

## Education

```python
edu = pd.read_csv('Edu/ACSST5Y2021.S1501-Data.csv')
#S1501_C01_006E: population over 25 years
#S1501_C01_014E: over 25 years, high school graduate or higher
#S1501_C01_015E: over 25 years, bachelor's degree or higher
edu = edu[['NAME','S1501_C01_006E','S1501_C01_015E']].rename(columns=
            {'S1501_C01_006E':'pop_over25','S1501_C01_015E':'bachelor_higher'})
edu.head()
```

```python
edu['NAME'] = edu['NAME'].apply(lambda x: x.split()[1])
edu = edu.rename(columns = {'NAME':'ZIP'})
edu.head()
```

```python
edu['ZIP'] = edu['ZIP'].astype(int)
```

```python
#manhattan avg
Manhattan = edu.loc[(edu['ZIP']>10000) & (edu['ZIP']<10300)]
Manhattan = Manhattan.loc[Manhattan['pop_over25']!=0]
Man_pop = int(Manhattan['pop_over25'].mean())
Man_ba = int(Manhattan['bachelor_higher'].mean())
print(Man_pop)
print(Man_ba)
```

```python
#Staten Island avg
Staten = edu.loc[(edu['ZIP']>10300) & (edu['ZIP']<10400)]
Staten = Staten.loc[Staten['pop_over25']!=0]
SI_pop = int(Staten['pop_over25'].mean())
SI_ba = int(Staten['bachelor_higher'].mean())
print(SI_pop)
print(SI_ba)
```

```python
#Bronx Island avg
Bronx = edu.loc[(edu['ZIP']>10400) & (edu['ZIP']<10500)]
Bronx = Bronx.loc[Bronx['pop_over25']!=0]
Br_pop = int(Bronx['pop_over25'].mean())
Br_ba = int(Bronx['bachelor_higher'].mean())
print(Br_pop)
print(Br_ba)
```

```python
#BK avg
king = edu.loc[(edu['ZIP']>=11200) & (edu['ZIP']<11300)]
king = king.loc[king['pop_over25']!=0]
BK_pop = int(king['pop_over25'].mean())
BK_ba = int(king['bachelor_higher'].mean())
print(BK_pop)
print(BK_ba)
```

```python
#queens
queens = edu.loc[((edu['ZIP']>=11000) & (edu['ZIP']<11200)) | ((edu['ZIP']>=11300) & (edu['ZIP']<11700))]
queens = queens.loc[queens['pop_over25']!=0]
Q_pop = int(queens['pop_over25'].mean())
Q_ba = int(queens['bachelor_higher'].mean())
print(Q_pop)
print(Q_ba)
```

```python
for i in edu.loc[edu['pop_over25']==0].index:
    if edu['ZIP'].iloc[i] > 10000 and edu['ZIP'].iloc[i] < 10300:
        edu['pop_over25'].iloc[i] = Man_pop
        edu['bachelor_higher'].iloc[i] = Man_ba
    elif edu['ZIP'].iloc[i] > 10300 and edu['ZIP'].iloc[i] < 10400:
        edu['pop_over25'].iloc[i] = SI_pop
        edu['bachelor_higher'].iloc[i] = SI_ba
    elif edu['ZIP'].iloc[i] > 10400 and edu['ZIP'].iloc[i] < 10500:
        edu['pop_over25'].iloc[i] = Br_pop
        edu['bachelor_higher'].iloc[i] = Br_ba
    elif edu['ZIP'].iloc[i] > 11200 and edu['ZIP'].iloc[i] < 11300:
        edu['pop_over25'].iloc[i] = BK_pop
        edu['bachelor_higher'].iloc[i] = BK_ba
    else:
        edu['pop_over25'].iloc[i] = Q_pop
        edu['bachelor_higher'].iloc[i] = Q_ba
```

```python
edu['prop_bachelor'] = edu['bachelor_higher']/edu['pop_over25']
```
